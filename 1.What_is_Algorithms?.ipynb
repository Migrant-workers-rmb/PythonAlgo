```python
def some_func_to_sum(n):
    final_result=0
    for x in range(n+1):
        final_result += x
    return final_result

some_func_to_sum(5)
```

```python
def another_func_to_sum(n):
    final_result=(1+n)*n*0.5
    return final_result

another_func_to_sum(5)
```

Big-o :1 constant <br>

Big-o :n^2 Quadratic <br>
Big-o :n^3 Cubic <br>
Big-o :2^n Exponential <br>




```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

n = np.linspace(1,100)
labels = ['Constant','Logarithmic','Linear','Log Linear','Quadratic','Cubic','Exponential']

big_o=[np.ones(n.shape),np.log(n),n,n*np.log(n),n**2,n**3,2**n]

plt.figure(figsize=(20,10))
plt.ylim(0,50)

for i in range(len(big_o)):
    plt.plot(n,big_o[i], label = labels[i])

plt.legend(loc=0)
plt.ylabel('(Relative) Runtime for Comparison')
plt.xlabel('n')

```
